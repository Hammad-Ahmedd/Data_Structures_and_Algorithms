### ğŸ’» Clean C++ Implementation

```cpp
#include <iostream>
#include <vector>
#include <climits>   // For INT_MIN
#include <algorithm> // For std::max

int kadanes(std::vector<int>& nums) {
    int bestSum = INT_MIN; // Overall record
    int currSum = 0;       // Current running total

    for (int x : nums) {
        currSum += x;
        
        // 1. Update the record holder
        bestSum = std::max(bestSum, currSum);
        
        // 2. If it's a burden (negative), start fresh
        if (currSum < 0) {
            currSum = 0;
        }
    }
    
    return bestSum;
}

int main() {
    std::vector<int> arr = {-2, 1, -3, 4, -1, 2, 1, -5, 4};
    std::cout << "Max Subarray Sum: " << kadanes(arr) << std::endl;
    return 0;
}
```
